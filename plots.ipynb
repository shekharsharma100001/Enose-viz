```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'e-nose_dataset_12_beef_cuts.xlsx'

# Load all sheet names
sheet_names = pd.ExcelFile(file_path).sheet_names

df = pd.read_excel(file_path, sheet_name=sheet_names[0])
x = df['TVC']
y1 = df['MQ135']
y2 = df['MQ136']
y3 = df['MQ137']
y4 = df['MQ138']
y5 = df['MQ2']
y6 = df['MQ3']
y7 = df['MQ4']
y8 = df['MQ5']
y9 = df['MQ6']
y10 = df['MQ8']
y11 = df['MQ9']
y_values = [y1, y2, y3, y4 ,y5 ,y6, y7, y8, y9, y10, y11]


# Create 4x3 subplots
fig, axes = plt.subplots(4, 3, figsize=(14, 10))
axes = axes.flatten()

# Plot each y series
for i in range(11):
    axes[i].plot(x, y_values[i], label=f"Line {i+1}")
    axes[i].set_title(f"Line Chart {i+1}")
    axes[i].legend()

# Hide the unused 12th subplot
axes[11].axis('off')

plt.tight_layout()
plt.show()

```
